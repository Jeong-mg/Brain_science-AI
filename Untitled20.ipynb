```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn. linear_model import LogisticRegression

from tensorflow.keras.models import Sequential
from tensorflow.keras. layers import Dense
from tensorflow.keras.optimizers import SGD

from draw_plot_2 import plot_decision_boundary, draw_data
```

```python
from sklearn.datasets import load_iris, load_diabetes
import tensorflow as tf

# 데이터
diabetes_data = load_diabetes()
X_data = diabetes_data.data[:, 3]
y_data = diabetes_data.target

# 모델 정의
model = tf.keras.Sequential ([
  tf.keras. layers. Dense(1, input_dim=1)
])

# 모델 컴파일
model. compile(optimizer='sgd', loss='mean_squared_error')

# 모델 학습
model.fit(X_data, y_data, epochs=100, verbose=0)

# 예촉
print(model.predict([10]))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# 데이터
diabetes_data = load_diabetes()
X_data = diabetes_data.data[:, 3]
y_data = diabetes_data.target
X_data = torch.tensor(X_data.reshape(-1, 1)). float()
y_data = torch.tensor(y_data.reshape(-1, 1)). float()

# 모델 정의
class LinearRegressionModel (nn. Module) :
  def __init__ (self):
    super(LinearRegressionModel, self).__init__()
    self.linear = nn.Linear(1, 1)

  def forward(self, x):
    return self.linear(x)

model = LinearRegressionModel()

# 손실 함수와 최적화 정의
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# 모델 학습
for epoch in range(100):
  optimizer.zero_grad()
  outputs = model (X_data)
  loss = criterion(outputs, y_data)
  loss. backward()
  optimizer. step()

# 예촉
print (model(torch. tensor([[10.]])))
```

```python
# AND 데이터
X_and = np.array([[0, 0], [0, 5], [5, 0], [5, 5]])
y_and = np.array([0, 0, 0, 1])

# OR 데이터
X_or = np.array( [[0, 0], [0, 5], [5, 0], [5, 5]])
y_or = np.array([0, 1, 1, 1])

# XOR 데이터
X_xor = np.array([[0, 0], [0, 5], [5, 0], [5, 5]])
y_xor = np.array([0, 1, 1, 0])

draw_data(X_and, y_and)
draw_data(X_or, y_or)
draw_data(X_xor, y_xor)
```

```python
# 로지스틱 회귀 모델 학습
model_logistic_and = LogisticRegression(max_iter=100)
model_logistic_and. fit(X_and, y_and)

# 로지스틱 회귀 모델 학습
model_logistic_or = LogisticRegression(max_iter=100)
model_logistic_or.fit(X_or, y_or)

model_logistic_xor = LogisticRegression(max_iter=100)
model_logistic_xor. fit(X_xor, y_xor)

# 학습 과정 시각화
plot_decision_boundary(X_and, y_and, model_logistic_and, "Logistic Regression (AND)")
plot_decision_boundary(X_or, y_or, model_logistic_or, "Logistic Regression (OR)")
plot_decision_boundary(X_xor, y_xor, model_logistic_xor, "Logistic Regression (XOR)")
```

```python
# 모델 생성 및 훈련
model_and = Sequential ([
  Dense(1, activation='sigmoid')
])
model_and.compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy', metrics=['accuracy'])
model_and.fit(X_and, y_and, epochs=100, verbose=0)

model_or = Sequential ([
  Dense(1, activation='sigmoid')
])
model_or. compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy', metrics=['accuracy'])
model_or.fit(X_or, y_or, epochs=100, verbose=0)

model_xor = Sequential ([
  Dense(1, activation='sigmoid')
])
model_xor. compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy', metrics=['accuracy'])
model_xor. fit(X_xor, y_xor, epochs=100, verbose=0)

# 학습 과정 시각화
plot_decision_boundary(X_and, y_and, model_and, "Logistic Regression (AND)")
plot_decision_boundary(X_or, y_or, model_or, "Logistic Regression (OR)")
plot_decision_boundary(X_xor, y_xor, model_xor, "Logistic Regression (XOR)")
```

```python
model_dnn = Sequential ([
  Dense(8, activation='relu', input_shape=(2,)),
  Dense(8, activation='relu'),
  Dense(1, activation='sigmoid')
])

model_dnn.compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy', metrics=['accuracy'])
model_dnn. fit(X_xor, y_xor, epochs=100, verbose=0)

plot_decision_boundary(X_xor, y_xor, model_dnn, "DNN (XOR)")
```
